```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("cleaned_employee_data.csv")
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.hist(df['Work_Hours'])
plt.title("Work Hours Distribution")
plt.xlabel("Work Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean Work Hours:", df['Work_Hours'].mean())
print("Median Work Hours:", df['Work_Hours'].median())
```

```python
print("Mean Productivity:", df['Productivity_Index'].mean())
print("Median Productivity:", df['Productivity_Index'].median())
```

```python
df['Work_Type'] = df['Work_Hours'].apply(lambda x: 'Normal' if x <= 40 else 'Overtime')
```

```python
df.head()
```

```python
df['Work_Type'].value_counts()
```

```python
df.boxplot(column='Productivity_Index', by='Work_Type')
plt.title("Productivity: Normal vs Overtime")
plt.suptitle("")
plt.xlabel("Work Type")
plt.ylabel("Productivity Index")
plt.show()
```

```python
plt.scatter(df['Work_Hours'], df['Productivity_Index'])
plt.title("Work Hours vs Productivity")
plt.xlabel("Work Hours")
plt.ylabel("Productivity Index")
plt.show()
```
